```python
# Remplazando valores N, S con No, Si en variable ESTU_PRIVADO_LIBERTAD y visualizando

df_datos["ESTU_PRIVADO_LIBERTAD"] = df_datos["ESTU_PRIVADO_LIBERTAD"].str.replace("Noo", "N")
df_datos["ESTU_PRIVADO_LIBERTAD"] = df_datos["ESTU_PRIVADO_LIBERTAD"].str.replace("No", "N")
df_datos["ESTU_PRIVADO_LIBERTAD"] = df_datos["ESTU_PRIVADO_LIBERTAD"].str.replace("Si", "S")
df_datos["ESTU_PRIVADO_LIBERTAD"] = df_datos["ESTU_PRIVADO_LIBERTAD"].str.replace("N", "No")
df_datos["ESTU_PRIVADO_LIBERTAD"] = df_datos["ESTU_PRIVADO_LIBERTAD"].str.replace("S", "Si")
df_datos.head()
```

```python
# Separando variable objetivo de predictores

Y = df_datos.RENDIMIENTO_GLOBAL

X = df_datos.drop(["RENDIMIENTO_GLOBAL"], axis = 1)
```

```python
# Importando herramienta de libreria y definiendo un SimpleImputer con estrategia el valor mas frecuente

from sklearn.impute import SimpleImputer

my_Imputer = SimpleImputer(strategy = "most_frequent")
```

```python
# Importando herramienta de libreria y realizando tranformacion de valores en columnas para eliminar datos faltantes en dataframe para prediciones

from sklearn.compose import make_column_transformer

ct = make_column_transformer((my_Imputer, ["ESTU_VALORMATRICULAUNIVERSIDAD"]),
                             (my_Imputer, ["ESTU_HORASSEMANATRABAJA"]),
                             (my_Imputer, ["FAMI_ESTRATOVIVIENDA"]),
                             (my_Imputer, ["FAMI_TIENEINTERNET"]),
                             (my_Imputer, ["FAMI_EDUCACIONPADRE"]),
                             (my_Imputer, ["FAMI_TIENELAVADORA"]),
                             (my_Imputer, ["FAMI_TIENEAUTOMOVIL"]),
                             (my_Imputer, ["ESTU_PAGOMATRICULAPROPIO"]),
                             (my_Imputer, ["FAMI_TIENECOMPUTADOR"]),
                             (my_Imputer, ["FAMI_EDUCACIONMADRE"]), remainder="passthrough")
```

```python
# Visualizando y realizando ajustes a salida de tranformacion de valores en columnas para eliminar datos faltantes

ct.set_output(transform="pandas")

X = ct.fit_transform(X)

X
```

```python
# Ajustando nombre de columnas trasformadas y organizando en dataframe de predictores

X = X.rename(columns={'simpleimputer-1__ESTU_VALORMATRICULAUNIVERSIDAD': 'ESTU_VALORMATRICULAUNIVERSIDAD',
                      'simpleimputer-2__ESTU_HORASSEMANATRABAJA': 'ESTU_HORASSEMANATRABAJA',
                      'simpleimputer-3__FAMI_ESTRATOVIVIENDA': 'FAMI_ESTRATOVIVIENDA',
                      'simpleimputer-4__FAMI_TIENEINTERNET': 'FAMI_TIENEINTERNET',
                      'simpleimputer-5__FAMI_EDUCACIONPADRE': 'FAMI_EDUCACIONPADRE',
                      'simpleimputer-6__FAMI_TIENELAVADORA': 'FAMI_TIENELAVADORA',
                      'simpleimputer-7__FAMI_TIENEAUTOMOVIL': 'FAMI_TIENEAUTOMOVIL',
                      'simpleimputer-8__ESTU_PAGOMATRICULAPROPIO': 'ESTU_PAGOMATRICULAPROPIO',
                      'simpleimputer-9__FAMI_TIENECOMPUTADOR': 'FAMI_TIENECOMPUTADOR',
                      'simpleimputer-10__FAMI_EDUCACIONMADRE': 'FAMI_EDUCACIONMADRE',
                      'remainder__ID': 'ID',
                      'remainder__PERIODO': 'PERIODO',
                      'remainder__ESTU_PRGM_DEPARTAMENTO': 'ESTU_PRGM_DEPARTAMENTO',
                      'remainder__ESTU_PRIVADO_LIBERTAD': 'ESTU_PRIVADO_LIBERTAD',
                      'remainder__coef_1': 'coef_1',
                      'remainder__coef_2': 'coef_2',
                      'remainder__coef_3': 'coef_3',
                      'remainder__coef_4': 'coef_4'
                      })

# Definir el nuevo orden de las columnas
new_order = ['ID', 'PERIODO', 'ESTU_PRGM_DEPARTAMENTO',
             'ESTU_VALORMATRICULAUNIVERSIDAD', 'ESTU_HORASSEMANATRABAJA',
             'FAMI_ESTRATOVIVIENDA', 'FAMI_TIENEINTERNET', 'FAMI_EDUCACIONPADRE',
             'FAMI_TIENELAVADORA', 'FAMI_TIENEAUTOMOVIL', 'ESTU_PRIVADO_LIBERTAD',
             'ESTU_PAGOMATRICULAPROPIO', 'FAMI_TIENECOMPUTADOR',
             'FAMI_EDUCACIONMADRE', 'coef_1', 'coef_2',
             'coef_3', 'coef_4']

# Reordenar el DataFrame
X = X.reindex(columns=new_order)

X
```

```python
# Calcular la moda de la columna y reemplazando el valor específico por la moda en variable FAMI_ESTRATOVIVIENDA en dataframe de predictores

moda = X['FAMI_ESTRATOVIVIENDA'].mode()[0]

X['FAMI_ESTRATOVIVIENDA'] = X['FAMI_ESTRATOVIVIENDA'].replace('Sin Estrato', moda)
```

```python
# Calcular la moda de la columna y reemplazando el valor específico por la moda en variable FAMI_EDUCACIONPADRE en dataframe de predictores

moda = X['FAMI_EDUCACIONPADRE'].mode()[0]

X['FAMI_EDUCACIONPADRE'] = X['FAMI_EDUCACIONPADRE'].replace('No sabe', moda)
```

```python
# Calcular la moda de la columna y reemplazando el valor específico por la moda en variable FAMI_EDUCACIONMADRE en dataframe de predictores

moda = X['FAMI_EDUCACIONMADRE'].mode()[0]

X['FAMI_EDUCACIONMADRE'] = X['FAMI_EDUCACIONMADRE'].replace('No sabe', moda)
```

```python
# Cambiando de string a numero en variables categoricas para variable objetivo

RENDIMIENTO = {'alto': 3, 'bajo': 0, 'medio-alto': 2, 'medio-bajo': 1}
Y = Y.map(RENDIMIENTO)
```

```python
# Importando herramienta y dividiendo datos para entrenamiento y test

from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.9, test_size=0.1)
```

```python
# Importando herramientas, definiendo modelo RandomForestClassifier e identificando score

from sklearn.ensemble import RandomForestClassifier

RFC = RandomForestClassifier(n_estimators=700, max_depth=17)
RFC.fit(X_train, Y_train)
RFC.score(X_test, Y_test)
```
