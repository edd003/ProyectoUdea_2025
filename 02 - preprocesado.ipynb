```python
# Remplazando valores N, S con No, Si

df_datos["ESTU_PRIVADO_LIBERTAD"] = df_datos["ESTU_PRIVADO_LIBERTAD"].str.replace("Noo", "N")
df_datos["ESTU_PRIVADO_LIBERTAD"] = df_datos["ESTU_PRIVADO_LIBERTAD"].str.replace("No", "N")
df_datos["ESTU_PRIVADO_LIBERTAD"] = df_datos["ESTU_PRIVADO_LIBERTAD"].str.replace("Si", "S")
df_datos["ESTU_PRIVADO_LIBERTAD"] = df_datos["ESTU_PRIVADO_LIBERTAD"].str.replace("N", "No")
df_datos["ESTU_PRIVADO_LIBERTAD"] = df_datos["ESTU_PRIVADO_LIBERTAD"].str.replace("S", "Si")
df_datos.head()
```

```python
# Cambiando de string a numero en variables categoricas y binarias

# ESTU_PRGM_DEPARTAMENTO ['AMAZONAS' 'ANTIOQUIA' 'ARAUCA' 'ATLANTICO' 'BOGOTÁ' 'BOLIVAR' 'BOYACA'
#  'CALDAS' 'CAQUETA' 'CASANARE' 'CAUCA' 'CESAR' 'CHOCO' 'CORDOBA'
#  'CUNDINAMARCA' 'GUAVIARE' 'HUILA' 'LA GUAJIRA' 'MAGDALENA' 'META'
#  'NARIÑO' 'NORTE SANTANDER' 'PUTUMAYO' 'QUINDIO' 'RISARALDA' 'SAN ANDRES'
#  'SANTANDER' 'SUCRE' 'TOLIMA' 'VALLE' 'VAUPES']

DEPARTAMENTO = {'AMAZONAS':1, 'ANTIOQUIA':2, 'ARAUCA':3, 'ATLANTICO':4, 'BOGOTÁ':5, 'BOLIVAR':6, 'BOYACA':7,
 'CALDAS':8, 'CAQUETA':9, 'CASANARE':10, 'CAUCA':11, 'CESAR':12, 'CHOCO':13, 'CORDOBA':14,
 'CUNDINAMARCA':15, 'GUAVIARE':16, 'HUILA':17, 'LA GUAJIRA':18, 'MAGDALENA':19, 'META':20,
 'NARIÑO':21, 'NORTE SANTANDER':22, 'PUTUMAYO':23, 'QUINDIO':24, 'RISARALDA':25, 'SAN ANDRES':26,
 'SANTANDER':27, 'SUCRE':28, 'TOLIMA':29, 'VALLE':30, 'VAUPES':31}

df_datos["ESTU_PRGM_DEPARTAMENTO"]=df_datos["ESTU_PRGM_DEPARTAMENTO"].map(DEPARTAMENTO)


# ESTU_VALORMATRICULAUNIVERSIDAD ['Entre 1 millón y menos de 2.5 millones'
#  'Entre 2.5 millones y menos de 4 millones'
#  'Entre 4 millones y menos de 5.5 millones'
#  'Entre 5.5 millones y menos de 7 millones'
#  'Entre 500 mil y menos de 1 millón' 'Menos de 500 mil'
#  'Más de 7 millones' 'No pagó matrícula']

VALORMATRICULA = {'Entre 1 millón y menos de 2.5 millones':2.49,
 'Entre 2.5 millones y menos de 4 millones':3.99,
 'Entre 4 millones y menos de 5.5 millones':5.49,
 'Entre 5.5 millones y menos de 7 millones':6.99, 'Entre 500 mil y menos de 1 millón':0.99, 'Menos de 500 mil':0.49,
 'Más de 7 millones':7, 'No pagó matrícula':0}

df_datos["ESTU_VALORMATRICULAUNIVERSIDAD"]=df_datos["ESTU_VALORMATRICULAUNIVERSIDAD"].map(VALORMATRICULA)


# ESTU_HORASSEMANATRABAJA ['0' 'Entre 11 y 20 horas' 'Entre 21 y 30 horas' 'Menos de 10 horas'
#  'Más de 30 horas']

HORASSEMANA = {'0':0, 'Entre 11 y 20 horas':20, 'Entre 21 y 30 horas':30, 'Menos de 10 horas':9,
 'Más de 30 horas':31}

df_datos["ESTU_HORASSEMANATRABAJA"]=df_datos["ESTU_HORASSEMANATRABAJA"].map(HORASSEMANA)


# FAMI_ESTRATOVIVIENDA ['Estrato 1' 'Estrato 2' 'Estrato 3' 'Estrato 4' 'Estrato 5' 'Estrato 6'
#  'Sin Estrato']

ESTRATO = {'Estrato 1':1, 'Estrato 2':2, 'Estrato 3':3, 'Estrato 4':4, 'Estrato 5':5, 'Estrato 6':6, 'Sin Estrato':1}

df_datos["FAMI_ESTRATOVIVIENDA"]=df_datos["FAMI_ESTRATOVIVIENDA"].map(ESTRATO)


# FAMI_TIENEINTERNET ['No' 'Si']

INTERNET = {'No':0, 'Si':1}
df_datos["FAMI_TIENEINTERNET"]=df_datos["FAMI_TIENEINTERNET"].map(INTERNET)


# FAMI_EDUCACIONPADRE ['Educación profesional completa' 'Educación profesional incompleta'
#  'Ninguno' 'No Aplica' 'No sabe' 'Postgrado' 'Primaria completa'
#  'Primaria incompleta' 'Secundaria (Bachillerato) completa'
#  'Secundaria (Bachillerato) incompleta' 'Técnica o tecnológica completa'
#  'Técnica o tecnológica incompleta']

EDUCACIONPADRE = {'Educación profesional completa':8, 'Educación profesional incompleta':7,
 'Ninguno':0, 'Postgrado':9, 'Primaria completa':2,
 'Primaria incompleta':1, 'Secundaria (Bachillerato) completa':4,
 'Secundaria (Bachillerato) incompleta':3, 'Técnica o tecnológica completa':6,
 'Técnica o tecnológica incompleta':5, 'No Aplica':0, 'No sabe':0}

df_datos["FAMI_EDUCACIONPADRE"]=df_datos["FAMI_EDUCACIONPADRE"].map(EDUCACIONPADRE)


#FAMI_TIENELAVADORA ['No' 'Si']

TIENELAVADORA = {'No':0, 'Si':1}
df_datos["FAMI_TIENELAVADORA"]=df_datos["FAMI_TIENELAVADORA"].map(TIENELAVADORA)


#FAMI_TIENEAUTOMOVIL ['No' 'Si']

TIENEAUTOMOVIL = {'No':0, 'Si':1}
df_datos["FAMI_TIENEAUTOMOVIL"]=df_datos["FAMI_TIENEAUTOMOVIL"].map(TIENEAUTOMOVIL)


#ESTU_PRIVADO_LIBERTAD ['N0' 'Si']

ESTU_PRIVADO = {'No':0, 'Si':1}
df_datos["ESTU_PRIVADO_LIBERTAD"]=df_datos["ESTU_PRIVADO_LIBERTAD"].map(ESTU_PRIVADO)


#ESTU_PAGOMATRICULAPROPIO ['No' 'Si']

PAGOMATRICULA = {'No':0, 'Si':1}
df_datos["ESTU_PAGOMATRICULAPROPIO"]=df_datos["ESTU_PAGOMATRICULAPROPIO"].map(PAGOMATRICULA)


#FAMI_TIENECOMPUTADOR ['No' 'Si']

TIENECOMPUTADOR = {'No':0, 'Si':1}
df_datos["FAMI_TIENECOMPUTADOR"]=df_datos["FAMI_TIENECOMPUTADOR"].map(TIENECOMPUTADOR)


# FAMI_EDUCACIONMADRE ['Educación profesional completa' 'Educación profesional incompleta'
#  'Ninguno' 'No Aplica' 'No sabe' 'Postgrado' 'Primaria completa'
#  'Primaria incompleta' 'Secundaria (Bachillerato) completa'
#  'Secundaria (Bachillerato) incompleta' 'Técnica o tecnológica completa'
#  'Técnica o tecnológica incompleta']

EDUCACIONMADRE = {'Educación profesional completa':8, 'Educación profesional incompleta':7,
 'Ninguno':0, 'Postgrado':9, 'Primaria completa':2,
 'Primaria incompleta':1, 'Secundaria (Bachillerato) completa':4,
 'Secundaria (Bachillerato) incompleta':3, 'Técnica o tecnológica completa':6,
 'Técnica o tecnológica incompleta':5, 'No Aplica':0, 'No sabe':0}

df_datos["FAMI_EDUCACIONMADRE"]=df_datos["FAMI_EDUCACIONMADRE"].map(EDUCACIONMADRE)


# RENDIMIENTO_GLOBAL ['alto' 'bajo' 'medio-alto' 'medio-bajo']

RENDIMIENTO = {'alto':4, 'bajo':1, 'medio-alto':3, 'medio-bajo':2}
df_datos["RENDIMIENTO_GLOBAL"]=df_datos["RENDIMIENTO_GLOBAL"].map(RENDIMIENTO)
```
